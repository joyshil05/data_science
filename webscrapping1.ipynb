```python
data=[]
for p in range(1,6):
    url = f'https://pharmeasy.in/api/otc/getCategoryProducts?categoryId=860&page={p}'
    reques = requests.get(url)
    joy=reques.json()
    for i in range(12):
        d = {
            'name':joy["data"]["products"][i]['name'],
            'brand':joy["data"]["products"][i]["manufacturer"],
            'mrpdecimal':joy["data"]["products"][i]['mrpDecimal'].split(".")[0] + " tk",
            "prodtype":joy["data"]["products"][i]['productType'],
            "salesprice":joy["data"]["products"][i]['salePriceDecimal'],
            "discount":joy["data"]["products"][i]['discountDecimal'],
            "discountpercent":joy["data"]["products"][i]['discountPercent']
            }
        print(d)
        data.append(d)
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
df.to_csv('data.csv',index=False)
```
